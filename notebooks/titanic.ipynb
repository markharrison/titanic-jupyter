```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('../data/titanic.csv')

# Remove rows with missing age
xf = df.dropna(subset=['age'])
 
# Remove rows with missing age and create a copy of the DataFrame
xf = df.dropna(subset=['age']).copy()
 
# Replace the 'survived' column with more descriptive labels
xf['survived'] = xf['survived'].replace({0: 'Died', 1: 'Survived'})

# Create a bar plot of the passenger class and survival 
survival_count = xf.groupby(['pclass', 'survived']).size().reset_index(name='counts')
sns.barplot(x='pclass', y='counts', hue='survived', data=survival_count)
plt.show()


 




```
